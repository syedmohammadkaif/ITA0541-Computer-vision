```python

import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# Upload image
file = list(files.upload().keys())[0]

# Read image
img = cv2.imread(file)

# Convert BGR to RGB
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Image size
rows, cols = img.shape[:2]

# Source points
pts1 = np.float32([[50,50], [300,50], [50,300], [300,300]])

# Destination points
pts2 = np.float32([[10,100], [300,50], [100,300], [300,250]])

# Perspective matrix
matrix = cv2.getPerspectiveTransform(pts1, pts2)

# Apply transformation
perspective = cv2.warpPerspective(img, matrix, (cols, rows))

# Display
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.title("Original Image")
plt.imshow(img)
plt.axis('off')

plt.subplot(1,2,2)
plt.title("Perspective Transformation")
plt.imshow(perspective)
plt.axis('off')

plt.show()
```
