```python

import cv2
import matplotlib.pyplot as plt
from google.colab import files

# Upload image
file = list(files.upload().keys())[0]

# Read image
img = cv2.imread(file)

# Convert BGR to RGB
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Original size
h, w = img.shape[:2]

# Resize images
bigger = cv2.resize(img, (w*2, h*2))
smaller = cv2.resize(img, (w//2, h//2))

# Display
plt.figure(figsize=(15,8))

# Original
plt.subplot(1,3,1)
plt.title(f"Original\n{w} x {h}")
plt.imshow(img)
plt.axis('off')

# Bigger Image
plt.subplot(1,3,2)
plt.title(f"Bigger Image\n{w*2} x {h*2}")
plt.imshow(bigger)
plt.axis('off')

# Smaller Image
plt.subplot(1,3,3)
plt.title(f"Smaller Image\n{w//2} x {h//2}")
plt.imshow(smaller)
plt.axis('off')

plt.show()

# Print sizes clearly
print("Original Size :", w, "x", h)
print("Bigger Size   :", w*2, "x", h*2)
print("Smaller Size  :", w//2, "x", h//2)
```
