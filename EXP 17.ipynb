```python

import cv2
import matplotlib.pyplot as plt
from google.colab import files

# Upload image
file = list(files.upload().keys())[0]

# Read image
img = cv2.imread(file)

# Convert BGR to RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Copy image
watermarked = img_rgb.copy()

# Add watermark text
cv2.putText(
    watermarked,
    'WATERMARK',
    (50, 50),                    # Position
    cv2.FONT_HERSHEY_SIMPLEX,
    1.5,                         # Font size
    (255, 0, 0),                 # Color
    3,                           # Thickness
    cv2.LINE_AA
)

# Display images
plt.figure(figsize=(10,5))

# Original Image
plt.subplot(1,2,1)
plt.title("Original Image")
plt.imshow(img_rgb)
plt.axis('off')

# Watermarked Image
plt.subplot(1,2,2)
plt.title("Watermarked Image")
plt.imshow(watermarked)
plt.axis('off')

plt.show()
```
