```python
import cv2
import matplotlib.pyplot as plt
from google.colab import files

# Upload image
file = list(files.upload().keys())[0]

# Read image
img = cv2.imread(file)

# Convert BGR to RGB
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Select ROI (crop)
roi = img[50:200, 50:200]

# Copy and paste ROI
img_copy = img.copy()
# Adjust the target coordinates for pasting the ROI to fit within the image
# The ROI is 150x150. If img_copy is 225x225, starting at (70,70) works (70+150=220 < 225)
img_copy[70:220, 70:220] = roi

# Display
plt.figure(figsize=(12,5))

plt.subplot(1,3,1)
plt.title("Original Image")
plt.imshow(img)
plt.axis('off')

plt.subplot(1,3,2)
plt.title("Cropped ROI")
plt.imshow(roi)
plt.axis('off')

plt.subplot(1,3,3)
plt.title("Copied & Pasted ROI")
plt.imshow(img_copy)
plt.axis('off')

plt.show()
```
